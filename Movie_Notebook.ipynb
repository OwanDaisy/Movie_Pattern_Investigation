Therefore we will consider as duplicates, movies that have the same title, release year and duration

Let us verify duplicate entries to ensure we will be counting the exact number of movies in our database

```python
-- Selects the id, title, release_year, country, duration, language, certification, gross, and budget
-- from the cinema.films table where the combination of title, release_year, country, and language
-- appears more than once in the table.

SELECT id, title, release_year, country, duration, language, certification, gross, budget
FROM cinema.films
WHERE (title, release_year,duration) IN (
    -- Subquery to find combinations of title, release_year, duration
    -- that appear more than once in the cinema.films table.
    SELECT title, release_year, duration
    FROM cinema.films
    GROUP BY title, release_year,duration
    HAVING COUNT(*) > 1
);
```

As suspected, there are indeed multiple movies repeating themselves. There are 237 unique movies along with their duplicates in total.

To obtain the actual count of distinct movies, let us query our database and include a clause to handle these duplicates.

```python
SELECT COUNT(*)
FROM (
    SELECT DISTINCT title, release_year, language
    FROM cinema.films
) AS distinct_films;
```

**Answer** : There are 4846 movies

### 2. There seems to be a lot of missing data in the gross and budget columns. How many rows have missing data? What would you recommend your manager to do with these rows?

There is a lot of missing data in the gross and budget columns.....but is it only these columns that contain missing data though? to accurately respond to this question, we will query our data for two answers :

- How many rows have missing data in the table?
- How many rows have missing data in the gross and budget columns?

```python
-- Querying for rows with missing data in the entire table
SELECT COUNT(*)
FROM cinema.films
WHERE id IS NULL
   OR title IS NULL
   OR release_year IS NULL
   OR country IS NULL
   OR duration IS NULL
   OR language IS NULL
   OR certification IS NULL
   OR gross IS NULL
   OR budget IS NULL;
```

```python
-- Querying for rows with missing data in the gross and budget columns
-- Select the count of rows with missing gross or budget values
SELECT COUNT(id) as rows_with_missing_gross_or_budget
FROM cinema.films
-- Filter rows where either gross or budget is null
WHERE gross IS null OR budget IS null;
```

**Answer** : We have 1,077 rows with missing data in the table, but 1,076 rows with missing data in the gross and budget columns combined. Let us keep in mind that this count takes into account duplicates as well. 

   Let us further our analysis by verifying the count of rows with missing data from a particular column only. If only a certain column shows a significant count of missing data, then we know how and where to focus our attention for addressing missing data.
  

```python
-- Select the count of rows where the 'gross' column is missing (null)
SELECT COUNT(id) as rows_with_missing_gross
FROM cinema.films
WHERE gross IS null;
```

```python
-- Select the count of rows where the budget column is null
SELECT COUNT(id) as rows_with_missing_budget
FROM cinema.films
WHERE budget IS null;
```

```python
-- Select the count of rows where both 'gross' and 'budget' columns are NULL
SELECT COUNT(id) as rows_with_missing_gross_and_budget
FROM cinema.films
WHERE gross IS null AND budget IS null;
```

In summary, we find there are 810 rows without a gross value, 430 rows without a budget value, and 164 rows missing both a budget and gross value. In total, there are 1,076 rows (810 + 430 - 164).

These are significant numbers, with the gross column having the most missing data.

### 5. What is the average duration of English versus French movies? 

Here, we handle duplicates to ensure we have a response as accurate as possible

**General Scope**

```python
-- Select the language and the average duration of films
SELECT language, AVG(duration) AS average_duration
-- From the films table in the cinema schema
FROM cinema.films
-- Filter the results to include only films in English or French
WHERE (language = 'English' AND (title, release_year, duration) NOT IN (
    -- Subquery to find combinations of title, release_year, duration
    -- that appear more than once in the cinema.films table.
    SELECT title, release_year, duration
    FROM cinema.films
    GROUP BY title, release_year, duration
    HAVING COUNT(*) > 1
))OR (language = 'French'
AND (title, release_year, duration) NOT IN (
    -- Subquery to find combinations of title, release_year, duration
    -- that appear more than once in the cinema.films table.
    SELECT title, release_year, duration
    FROM cinema.films
    GROUP BY title, release_year, duration
    HAVING COUNT(*) > 1
))
-- Group the results by language to calculate the average duration for each language
GROUP BY language;
```

**Movies Released Before 1997**

6.1  Which rating do most movies have?

```python
-- Select the certification and count of films for each certification
SELECT certification, COUNT(*) AS rating_count
-- From the films table in the cinema schema
FROM cinema.films
-- Filter the films to include only those released before 1997
WHERE release_year < 1997
-- Group the results by the certification column
GROUP BY cinema.films.certification
-- Order the results by the count of films in descending order
ORDER BY rating_count DESC;
```

The majority of the movies are rated 'R'. Let us verify the number of English movies with this rating.

```python
-- Select the count of films and their language
SELECT COUNT(*), language
-- From the cinema.films table
FROM cinema.films
-- Where the certification is 'R', the language is 'English', and the release year is before 1997
WHERE certification = 'R' AND language = 'English' AND release_year < 1997
-- Group the results by language
GROUP BY language;
```

About 96% of English movies are rated R

6.2 Let us visualise the profit margin for top movies

```python
-- Select the country, language, gross, budget, and calculated profit for films
SELECT title, country, language, gross, budget, (gross - budget) AS profit
-- From the cinema.films table
FROM cinema.films
-- Only include rows where gross and budget are not null and the release year is before 1997
WHERE gross IS NOT null AND budget IS NOT null AND release_year < 1997 AND id NOT IN (
    -- Subquery to find IDs of films that have duplicate combinations of country, language, gross, budget
    SELECT MIN(id)
    FROM cinema.films
    GROUP BY country, language, gross, budget
    HAVING COUNT(*) > 1
)
-- Order the results by profit in descending order
ORDER BY profit DESC;
```

The top 10 movies with the highest revenues are all in English and produced in the USA. This observation suggests that USA-produced movies may be more appealing and have a larger, more active audience than those from other countries.

Now, Instead of visualizing the amount of profit per movie, let us examine the percentage profit to obtain a more realistic view of their earnings compared to other movies. A large profit could mislead us into thinking a movie is more successful than others, whereas it might be due to a large budget.


```python
SELECT 
    id, 
    title, 
    release_year, 
    country, 
    duration, 
    language, 
    certification, 
    gross, 
    budget, 
    (gross - budget) * 100.0 / budget AS percentage_profit
FROM 
    cinema.films
WHERE 
    release_year < 1997
    AND gross IS NOT NULL
    AND budget IS NOT NULL
	AND (title, release_year, duration) NOT IN (
    -- Subquery to find combinations of title, release_year, duration
    -- that appear more than once in the cinema.films table.
    SELECT title, release_year, duration
    FROM cinema.films
    GROUP BY title, release_year, duration
    HAVING COUNT(*) > 1
)
ORDER BY 
    percentage_profit DESC;
```

This observation, together with the previous queries, lead us to acknowledge that not only are the highest revenue-generating movies made in USA, but most of them are in English (with a single Spanish movie inserting itself in the rank of top 10).
